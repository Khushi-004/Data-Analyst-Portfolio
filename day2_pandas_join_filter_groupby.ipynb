```python
import pandas as pd
# Players ki table
players=pd.DataFrame({
    'PlayerID': [1,2,3,4,5],
    'Name': ['Virat','Rohit','Bumrah','Dhoni','Hardik'],
    'Team': ['RCB','MI','MI','CSK','MI'],
    'Age': [35,36,30,42,30]
})

# runs ki table
runs=pd.DataFrame({
    'PlayerID':[1,2,3,5,1,2],
    'Match':[1,1,1,1,2,2],
    'Runs':[80,45,0,30,100,60]
})

print("Players table")
print(players)

print("\n runs table")
print(runs)

#players+runs ko jodo (inner join)
df=pd.merge(players,runs,on='PlayerID',how='inner')
print("Task 1- Joined data")
print(df)

#runs>50 wale (filter)
print("\n Task 2- 50+ Runs wale:")
print(df[df['Runs']>50])

#Team wise total runs
print("\n Task 3- Team wise Total Runs:")
print(df.groupby('Team')['Runs'].sum())
```
